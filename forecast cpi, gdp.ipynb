```r
trend.interpolation<-function(data,n,m=3,eta=NULL,start=c(2013,1),from="2013-01-01",to="2019-01-01",title,dir=paste(disk,"",sep="")) {
  library(foreach)
  # data=as.data.frame(gdp_interpolated_updated_1);if(ncol(data)>1){data=data.frame(data[,2:ncol(data)])}else{data=data};n=nrow(data);m=3;eta=which(colnames(data)=="real.gdp",arr.ind=TRUE);from="2002-04-01";to="2019-07-01";title=paste("National accounts",".indicators.m.2002q2-2019q2",sep="")
  x<-as.data.frame(data)
  if(!is.numeric(n)){cat("Provide for n only numeric values!")}else{cat("N is provided.")};n=n
  if(!exists("m")){m=3}else{
    if(!is.numeric(m)){cat("Provide for m only numeric values!");m=3}else{if(m==3){freq=12}else{if(m==4){freq=4}else{freq=1}}}}
  if(!exists("eta")){eta=NULL}else{if(is.null(eta)){eta=NULL}else{eta=which(colnames(data)==eta,arr.ind=TRUE)}} # back the number of specified variable
  if(is.null(title)){title="Interpolated.data"}else{title=title}
  # n<-as.numeric(n);m<-as.numeric(m)
  n1<-nrow(x) # all rows in data.frame object even though with NA values
  #################################################
  ################  table  ########################
  #################################################
  x # is data.frame object
  index=unique(which((!is.na(x)&x>0),arr.ind=TRUE)[,2])  # choose non-empty columns
  prediction1<-data.frame(matrix(0,n1,length(index))) # empty table for recording in it the new results
  for(j in index){if(any(is.na(x[,j]))|is.null(x[,j])){"End running the code!";print(paste(colnames(x)[j],"is ommited."))}else{if(is.null(n)){n=length(na.omit(x[,j]))}else{n=n}
    prediction1[,which(j==index,arr.ind=TRUE)]<-as.numeric(x[1:n1,j]);idj=which(j==index,arr.ind=TRUE);colnames(prediction1)[idj]<-colnames(x)[idj]}}# correct!
  x.1<-ts(prediction1,start=start,frequency=freq)
  for(j in 1:ncol(x.1)){if(exists("n")==FALSE){n=length(na.omit(x[,j]))}else{n=n}
    assign(paste(".interpolated.series.",j,sep=""),splines::interpSpline(na.omit(x.1[,j])~ts(seq(1,n,length.out=n)),seq(1,n*m,length.out=n*m)/m))}
  for(j in 1:ncol(x.1)){if(is.null(n)){n=length(na.omit(x[,j]))}else{n=n}
    assign(paste(".prediction",j,sep=""),predict(get(paste(".interpolated.series.",j,sep="")),seq(1,n*m,length.out=n*m)/m)$y)}
  prediction<-foreach::foreach(j = 1:ncol(x.1),.combine=cbind)%do%(get(paste(".prediction",j,sep="")))
  prediction<-data.frame(prediction)
  colnames(prediction)<-colnames(x)[index];
  if(m==4){delta="quarters"}else{delta="months"}
  time<-seq.Date(as.Date(from),as.Date(to),by=delta)[1:nrow(prediction)]
  prediction<-cbind(time,prediction)
  if(!is.null(eta)){
    result=list(prediction.eta=get(paste("prediction"))[,which(colnames(x)[index]==colnames(x)[eta-1],arr.ind=TRUE)])}else{result=list(prediction.eta=get(paste("prediction")))}
  if(is.null(dir))setwd("C:/Users/1/OneDrive/Computer/I/")else{setwd(dir)}
  WriteXLS::WriteXLS(prediction,paste(title,"xlsx",sep="."),FreezeRow = 1,FreezeCol = 1)
  return(prediction)
}
```
